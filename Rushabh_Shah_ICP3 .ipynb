Replace the null values with the mean

```python
data.isnull()
col1=data['Duration'].mean();
col2=data['Pulse'].mean();
col3=data['Maxpulse'].mean();
col4=data['Calories'].mean();
data['Duration'].fillna(value=col1,inplace= True)
data['Pulse'].fillna(value=col2,inplace= True)
data['Maxpulse'].fillna(value=col3,inplace= True)
data['Calories'].fillna(value=col4,inplace= True)
data
```
